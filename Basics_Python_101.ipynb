# IF-statements
https://pythonspot.com/conditional-statements/

```python
a = "you are beautiful"
b = "you are awesome"
 
if a == b: # need to use 2 equal signs
    print ('equal')    
else:
    print ('not equal')
```

```python
a = 100
b = 10
if a < b:
   print ('a smaller than 10')
else:
   print ('a is bigger than 10 or equal')
```

```python
# create an if statement comparing two string values and print the result

x = "chicken pie"
y = "chinese food"
if x == y:
    print ("Chicken pie is chinese food")
else:
    print ("Chicken pie is not chinese food")
```

```python
#if two integers are the same, use elif command 

a = 10 
b = 10

if a < b:
    print ("b is greater than a")
elif a > b:
    print ("a is greater than b")
else: 
    print ("a is equal to b")
```

Lists can be thought of as a series of boxes. Each box has a different value. For example, the boxes created by demolist = ['life', 42, 'the universe', 6, 'and', 9] 

Each box is referenced by its number so the statement demolist[0] would get 'life', demolist[1] would get 42 and so on up to demolist[5] getting 9.
